```python
import numpy as np
from sklearn.cluster import KMeans
```

```python
d=np.array([[1,2],[2,1.3],[3,2],[2,2.5],[4,5],[5,6],[6,4.5],[7,6]])
```

```python
plt.scatter(d[:,0],d[:,1])
```

```python
algo=KMeans(2)
algo.fit(d)
```

```python
clus=algo.cluster_centers_
lab=algo.labels_
```

```python
m=['>','s','8','*']
for i in range(len(d)):
    plt.scatter(*d[i],marker=m[lab[i]],s=150)
plt.scatter(clus[:,0],clus[:,1],marker='X',s=200)
    
```

```python
class kmeans:
    def __init__(s,n=8,max_iter=300,tol=0.00001):
        s.n=n
        s.it=max_iter
        s.tol=tol
    def fit(s,data):
        s.clust=[]
        for i in range(s.n):
            s.clust.append(data[i])
        for i in range(s.it):
            dicts={}
            for j in range(s.n):
                dicts[j]=[]
            for pt in data:
                t=[np.linalg.norm(pt-c) for c in s.clust]
                dicts[t.index(min(t))].append(pt)
            prev=s.clust.copy()
            for k in dicts:
                s.clust[k]=np.mean(dicts[k],axis=0)
            if np.linalg.norm(np.array(prev)-np.array(s.clust))<s.tol:
                break
        return np.array(s.clust)
```

```python
a=kmeans(n=2)
a.fit(d)
```
